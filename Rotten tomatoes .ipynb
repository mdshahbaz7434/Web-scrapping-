```python
import requests 
from bs4 import BeautifulSoup
```

```python
base_site = "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/2/"
```

```python
response = requests.get(base_site)
response
```

```python
response.ok   # means all okay we can contniue our work 
```

```python
html = response.content 
```

```python
soup = BeautifulSoup(html,"lxml")  #lxml is showing poreper file 
```

```python
with open("Rotten_tomatoes_page_2_LXML_Parser.html","wb") as file:
    file.write(soup.prettify("utf-8"))
```

```python
divs = soup.find_all("div",{"class":"col-sm-18 col-full-xs countdown-item-content"})
```

```python
divs
```

### Now Extracting the tittle and year of each movie 

```python
headings = [div.find("h2") for div in divs]
headings
```

```python
headings[0]
```

### Title 

```python
movie_names = [heading.find("a").string for heading in headings]
movie_names
```

### Year

```python
years = [heading.find("span",class_ = "start-year").string for heading in headings]
```

```python
years
```

### removing brackets from year

```python
years[0].strip("()")
```

```python
years = [year.strip("()") for year in years]
years
```

```python
years = [int(year) for year in years]
years
```

### Critics 

```python
critics = [div.find("div",{"class":"info critics-consensus"}) for div in divs]
```

```python
critics
```

```python
critics[0]
```

```python
critics[0].contents[1].strip()
```

```python
consensus_text = [con.contents[1].strip() for con in critics]
consensus_text
```

### Directed by 

```python
directors = [div.find("div",class_ = "director") for div in divs]
```

```python
directors
```
